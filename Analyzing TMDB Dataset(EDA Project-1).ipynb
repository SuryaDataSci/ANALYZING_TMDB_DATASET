```python
#importing the necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset for analyzing the dataset
data=pd.read_csv(r'C:\Users\VENKATA SURYA\OneDrive\Desktop\everything about ai and ds\tmdb_movies_data.csv')
data.info()
```

```python
#describing i.e taking statistical apperance of dataset
data.describe()
```

```python
#dataset has null values so try to figure it out
data.isna().sum()
```

```python
#now filling the null values with fillna function
data.fillna(0)
print()
```

```python
#now removing duplicate rows
sum(data.duplicated())
```

```python
#now droping the duplicate rows
data.drop_duplicates(inplace=True)
print("after process completion",data.shape)
```

```python
#now checking the movies relaease time in the dataset to analyze further trend.
data['release_date']=pd.to_datetime(data['release_date'])
data['release_date'].head()
```

```python
#now we remove the unused column for the above dataset.
data.drop(['budget_adj','revenue_adj','overview','imdb_id','homepage','tagline'],axis =1,inplace = True)
print("Afetr Removing Unused Columns (Rows,Columns) : ",data.shape)
```

```python
#now drop the rows of unappropriate values
print("rows with zero values:",data[(data['budget']==0)].shape[0])
print("rows with zero in reverse:",data[(data['revenue']==0)].shape[0])
```

```python
#eda processes start
#now first finding the highest release of movies.
dd=data.groupby('release_year').count()['id']
print(dd.tail())
data.groupby('release_year').count()['id'].plot(xticks = np.arange(1960,2016,5))
sns.set(rc={'figure.figsize':(10,5)})
plt.title("year with high releases",fontsize=14)
plt.xlabel('release of year',fontsize=13)
plt.ylabel('number of movies',fontsize=13)
sns.set_style("whitegrid")
```

```python
#calculate todays profit
data['Profit']=data['revenue']-data['budget']
```

```python
#now we will findout the movies with highest and lowest profut
def find_minmax(x):
    min_index=data[x].idxmin()
    high_index=data[x].idxmax()
    high=pd.DataFrame(data.loc[high_index,:])
    low=pd.DataFrame(data.loc[min_index,:])
    print("movie with highest profit"+ x + ":",data['original_title'][high_index])
    print("movie with lowest profit"+ x + ":",data['original_title'][min_index])
    return pd.concat([high,low],axis=1)
find_minmax('Profit')
```

```python
#now plotting the graph foe most succesful and high profit movies
info=pd.DataFrame(data['Profit'].sort_values(ascending = False))
info['original_title']=data['original_title']
dd=list(map(str,(info['original_title'])))
x=list(dd[:10])
y=list(info['Profit'][:10])

ax=sns.pointplot(x=y,y=x)
sns.set(rc={'figure.figsize':(10,5)})
ax.set_title("most profitable movies",fontsize=15)
ax.set_xlabel("Profit",fontsize=13)
sns.set_style("darkgrid")
```

```python
#which year has highest prift rate combined
data.groupby('release_year')['Profit'].mean().plot()
plt.title("year with most profit",fontsize=14)
plt.xlabel('Release year',fontsize=13)
plt.ylabel('Average Profit',fontsize=13)
sns.set(rc={'figure.figsize':(10,5)})
sns.set_style("whitegrid")
```

```python
#relation between audience and movies of their popularity
data.groupby('runtime')['popularity'].mean().plot(figsize = (13,5),xticks=np.arange(0,1000,100))

plt.title("movies reached by popularity",fontsize=14)
plt.xlabel('Runtime',fontsize=13)
plt.ylabel('Average Popularity',fontsize=13)
sns.set(rc={'figure.figsize':(10,5)})
sns.set_style("whitegrid")
```

```python
ax = sns.regplot(x=data['revenue'], y=data['budget'],color='c')
ax.set_title("Revenue Vs Budget",fontsize=13)
ax.set_xlabel("Revenue",fontsize=12)
ax.set_ylabel("Budget",fontsize=12)
sns.set(rc={'figure.figsize':(6,4)})
sns.set_style("whitegrid")

#find the correlation between them
data['budget'] = data['budget'].replace(0,np.NAN)
data['revenue'] = data['revenue'].replace(0,np.NAN)
data['Profit'] = data['Profit'].replace(0,np.NAN)
data_corr = data.corr()
print("Correlation Between Revenue And Budget : ",data_corr.loc['revenue','budget'])
```

```python
#checking the average ratings of every movie
ax = sns.regplot(x=data['release_year'], y=data['vote_average'],color='c')
ax.set_title("Release Year Vs Vote Average",fontsize=13)
ax.set_xlabel("Release Year",fontsize=12)
ax.set_ylabel("Vote Average",fontsize=12)
sns.set(rc={'figure.figsize':(6,4)})
sns.set_style("whitegrid")
print("Correlation Between Release Year And Vote Average : ",data_corr.loc['release_year','vote_average'])
```

```python
#now which month has produced many movies with good gross profit
month_release = data['release_date'].dt.month
number_of_release = month_release.value_counts().sort_index()
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
number_of_release = pd.DataFrame(number_of_release)
number_of_release['month'] = months
number_of_release.rename(columns = {'release_date':'number_of_release'},inplace=True)
number_of_release.plot(x='month',kind='bar',fontsize = 11,figsize=(8,6))
plt.title('Months vs Number Of Movie Releases',fontsize = 15)
plt.xlabel('Month',fontsize = 13)
plt.ylabel('Number of movie releases',fontsize = 13)
sns.set_style("darkgrid")
```

```python
#now building and checking the relation between genre and month of movie that suits
def count_genre(x):
    data_plot = data[x].str.cat(sep = '|')
    dd = pd.Series(data_plot.split('|'))
    info = dd.value_counts(ascending=False)
    return info
total_genre_movies = count_genre('genres')
total_genre_movies.plot(kind= 'barh',figsize = (13,6),fontsize=12,colormap='tab20c')
plt.title("Genre With Highest Release",fontsize=15)
plt.xlabel('Number Of Movies',fontsize=13)
plt.ylabel("Genres",fontsize= 13)
sns.set_style("whitegrid")
```

```python
#now checking which genre has highest rated for a moviw to be based
i = 0
genre_count = []
for genre in total_genre_movies.index:
    genre_count.append([genre, total_genre_movies[i]])
    i = i+1
    
plt.rc('font', weight='bold')
f, ax = plt.subplots(figsize=(5, 5))
genre_count.sort(key = lambda x:x[1], reverse = True)
labels, sizes = zip(*genre_count)
labels_selected = [n if v > sum(sizes) * 0.01 else '' for n, v in genre_count]
ax.pie(sizes, labels=labels_selected,
       autopct = lambda x:'{:2.0f}%'.format(x) if x > 1 else '',
       shadow=False, startangle=0)
ax.axis('equal')
plt.tight_layout()
```

```python
#which genre suits which actor relation
count_actor_flims = count_genre('cast')
count_actor_flims.iloc[:20].plot.bar(figsize=(13,6),colormap= 'tab20c',fontsize=12)
plt.title("Most Frequent Actor",fontsize=15)
plt.xticks(rotation = 70)
plt.xlabel('Actor',fontsize=13)
plt.ylabel("Number Of Movies",fontsize= 13)
sns.set_style("whitegrid")
```

```python
#top companies with most amount of releases of movies
production_companies = count_genre('production_companies')
production_companies.iloc[:20].plot(kind='barh',figsize=(16,8),fontsize=13)
plt.title("Production Companies Vs Number Of Movies",fontsize=15)
plt.xlabel('Number Of Movies',fontsize=14)
sns.set_style("whitegrid")
```

```python
#which director of the product companies mainly relaese more movies
count_director_movies = count_genre('director')
count_director_movies.iloc[:20].plot(kind='bar',figsize=(13,6),fontsize=12)
plt.title("Director Vs Number Of Movies",fontsize=15)
plt.xticks(rotation=70)
plt.ylabel("Number Of Movies",fontsize= 13)
sns.set_style("whitegrid")
```
